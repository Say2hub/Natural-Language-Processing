```python
import nltk
from textblob import TextBlob
text = "Hello, I'm Sayantan Mukherjee. I currently Study at DJSCE. I Love Data Science,AI,ML Subjects and keen to learn more about them."
blob = TextBlob(text)

# Generate unigrams
unigrams = blob.ngrams(1)
print("Unigrams:", list(unigrams))

# Generate bigrams
bigrams = blob.ngrams(2)
print("Bigrams:", list(bigrams))

# Generate trigrams
trigrams = blob.ngrams(3)
print("Trigrams:", list(trigrams))
```

```python
import nltk
from nltk import FreqDist
from textblob import TextBlob

text = "Hello, I'm Sayantan Mukherjee. I currently Study at DJSCE. I Love Data Science,AI,ML Subjects and keen to learn more about them."
blob = TextBlob(text)

# Tokenize the text
tokens = blob.words
fdist = FreqDist(tokens)

# Print the frequency distribution
print("Frequency Distribution:")
print(fdist)

# Print the most common words
print("\nMost Common Words:")
print(fdist.most_common(5)) # Print the 5 most common words
```

```python
from collections import defaultdict, Counter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

story = """Once upon a time, in a faraway kingdom, there was a brave knight named Sir Arthur.
Sir Arthur had a loyal companion, a dragon named Drako. Together, they embarked on thrilling adventures to protect
the kingdom from evil forces. One day, they heard about a mysterious treasure hidden in the enchanted forest.
Determined to find it, Sir Arthur and Drako set off on their quest. As they journeyed through the forest, they
encountered magical creatures, solved ancient riddles, and faced dangerous challenges. Finally, after many trials,
they discovered the treasure and returned to the kingdom as heroes."""

words = story.split()

# Generate bigram counts
bigram_counts = defaultdict(Counter)
for i in range(len(words) - 1):
    bigram_counts[words[i]][words[i+1]] += 1

#Bigram probabilities
bigram_probs = defaultdict(dict)
for w1, next_words in bigram_counts.items():
    total_count = sum(next_words.values())
    for w2, count in next_words.items():
        bigram_probs[w1][w2] = count / total_count

# Predict next word given a current word and verify
predictions = []
for i in range(len(words) - 1):
    current_word = words[i]
    actual_next_word = words[i + 1]
    predicted_next_word = max(bigram_probs[current_word], key=bigram_probs[current_word].get) if current_word in bigram_probs else None
    predictions.append((current_word, actual_next_word, predicted_next_word))

# Print verification results
for current_word, actual_next_word, predicted_next_word in predictions:
    print(f"Current Word: '{current_word}' | Actual Next Word: '{actual_next_word}' | Predicted Next Word: '{predicted_next_word}' | Match: {actual_next_word == predicted_next_word}")

# Create a heatmap
bigram_df = pd.DataFrame(bigram_probs).fillna(0)
plt.figure(figsize=(14, 10))
sns.heatmap(bigram_df, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Probability'})
plt.title('Bigram Probabilities Heatmap')
plt.xlabel('Next Word')
plt.ylabel('Current Word')
plt.show()
```

```python
def predict_next_word_and_success_rate(story):
    words = story.split()
    bigram_counts = defaultdict(Counter)
    for i in range(len(words) - 1):
        bigram_counts[words[i]][words[i+1]] += 1

    bigram_probs = defaultdict(dict)
    for w1, next_words in bigram_counts.items():
        total_count = sum(next_words.values())
        for w2, count in next_words.items():
            bigram_probs[w1][w2] = count / total_count

    predictions = []
    correct_predictions = 0
    incorrect_predictions = 0
    total_predictions = 0

    for i in range(len(words) - 1):
        current_word = words[i]
        actual_next_word = words[i + 1]
        predicted_next_word = max(bigram_probs[current_word], key=bigram_probs[current_word].get) if current_word in bigram_probs else None
        is_correct = actual_next_word == predicted_next_word
        predictions.append((current_word, actual_next_word, predicted_next_word, is_correct))
        if is_correct:
            correct_predictions += 1
        else:
            incorrect_predictions += 1
        total_predictions += 1

    success_rate = (correct_predictions / total_predictions) * 100
    return predictions, success_rate, correct_predictions, incorrect_predictions, len(words)


predictions, success_rate, correct_predictions, incorrect_predictions, total_words = predict_next_word_and_success_rate(story)


print(f"\nSuccess Rate: {success_rate:.2f}%")
print(f"Number of Correct Predictions: {correct_predictions}")
print(f"Number of Incorrect Predictions: {incorrect_predictions}")
print(f"Total Number of Words: {total_words}")
```
